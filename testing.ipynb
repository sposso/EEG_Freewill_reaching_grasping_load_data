## Testing class to load EEG data

### Preprocessing

1. We downsample the data for the subjects with a sampling rate of 1000 Hz. Prior downsampling, an antialising lowpass infinitew impulse response filter with a 112.5 Hz cutoff was used, followed by a cubic spline interpolation.
2. After down sampling, we pandpass  filtered the data between 0.3 and 3 Hz using a zero-phase, 4th order Butterworth IIR filter.
This low-frequency range was selected because both foundational and recent studies have demonstrated that non-repetitive upper limb movements, including combined reach-and-grasp actions,are reliably represented within the low-frequency domain of the EEG. These slow cortical potentials, particularly movement-related cortical potentials (MRCPs), are highly informative for decoding motor intentions.

#### Studies

- **Unimanual and Bimanual Reach-and-Grasp Actions Can Be Decoded From Human EEG** : Features were extracted from the EEG within the 0.3–3 Hz frequency window during voluntary reach-and-grasp tasks.

- **Electroencephalographic Recording of the Movement-Related Cortical Potential During Motor Tasks**: Details the efficacy of bandpassing EEG in the <5 Hz range for capturing MRCPs during voluntary movements

- **Detection of Movement Related Cortical Potentials from EEG Using Convolutional Neural Networks**: Shows that a 0.1–3 Hz or 0.3–3 Hz window provides optimal extraction of premovement cortical signals for classification of upper limb actions

- **A Review of Techniques for Detection of Movement Intention Using EEG**: Reviews MRCP-based movement intention detection, emphasizing the utility of <5 Hz for decoding upper limb kinematics in neuroprosthesis control.
