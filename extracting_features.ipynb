```python
def FTA_features(trials, fs, max_freq):
    
    n_trials, n_channels, n_points = trials.shape
    
    # Frequency axis  and indices for positive frequencies up to max_freq
    freq_res = np.fft.fftfreq(n_points, d=1/fs) 
    pos_freqs = freq_res >= 0
    freques = freq_res[pos_freqs]
    freq_axis_indices= np.where(freques <= max_freq)[0]

    freq_axis_ = freques[freq_axis_indices]
    n_freq_components = len(freq_axis_indices)

    all_features = []

    real_features = []
    imag_features = []

    real_features_arr = np.zeros((n_trials, n_channels, n_freq_components))
    imag_features_arr = np.zeros((n_trials, n_channels, n_freq_components))

    for trial in range(n_trials):
        eeg_fragment = trials[trial, :, :]  # Current trial
        features = []
        real_feat = []
        imag_feat = []


        for ch in range(n_channels):
            fft_values = np.fft.fft(eeg_fragment[ch])  # Normalize the FFT output
            fft_values = fft_values[pos_freqs]  # Keep only positive frequencies
            amps = fft_values[freq_axis_indices]  # Keep only frequencies up to max_freq


            # dc component (real part only) + real and imaginary parts of other frequencies
            feat = [amps[0].real]
            r_feat = [amps[0].real]
            im_feat = [0.0]  # add DC imaginary explicitly
            real_features_arr[trial, ch, 0] = amps[0].real
            imag_features_arr[trial, ch, 0] = 0.0

            for a in amps[1:]:
                # Multiply by 2 to account for negative frequencies
                feat.extend([a.real*2, a.imag*2])
                r_feat.extend([a.real*2])
                im_feat.extend([a.imag*2])
                real_features_arr[trial, ch, len(r_feat)-1] = a.real*2
                imag_features_arr[trial, ch, len(im_feat)-1] = a.imag*2
            features.extend(feat)
            real_feat.extend(r_feat)
            imag_feat.extend(im_feat)

        all_features.append(features)
        real_features.append(real_feat)
        imag_features.append(imag_feat)
    
    real_features = np.array(real_features)  # Shape (n_trials, n_channels * n_freq_components)

    imag_features = np.array(imag_features)  # Shape (n_trials, n_channels  * n_freq_components)

    return np.array(all_features), real_features, imag_features, freq_axis_, n_freq_components, real_features_arr, imag_features_arr  
```

```python
# Example usage with a cosine signal with 500 sample points, sampled at 250 Hz
fs = 250  # Sampling frequency
max_freq = 5  # Maximum frequency to consider
n_samples = 500
t = np.arange(n_samples) / fs  # Time vector


signal = np.sin(2 * np.pi * 4* t)   # 4 Hz cosine signal

signal = signal.reshape(1, 1, -1)  # Reshape to (n_trials, n_channels, n_times)
print("Signal shape:", signal.shape)  # Should be (1, 1, 500)
features,real,ima,freq_axis, n_freq_comp,_,_ = FTA_features(signal, fs, max_freq)
print("FTA Features shape:", features.shape)  # Should be (1, number_of_features
```

```python
# plot all FTA features 
import matplotlib.pyplot as plt

# Plot original signal and its FTA features
fig, axes = plt.subplots(3, 1, figsize=(12, 8), constrained_layout=True)

# Original time-domain signal
axes[0].plot(t, signal[0, 0])
axes[0].set_title("Original Signal (Time Domain)")
axes[0].set_xlabel("Time (s)")
axes[0].set_ylabel("Amplitude")

# FTA features (already computed in 'features')
axes[1].stem(freq_axis, real[0], basefmt=" ")
axes[1].set_title("FTA Features")
axes[1].set_xlabel("Frequency (Hz)")
axes[1].set_ylabel("FT Real Amplitude")
axes[1].grid(True)
axes[1].set_xlim(0, 5)

# FTA features (already computed in 'features')
axes[2].stem(freq_axis, ima[0], basefmt=" ")
axes[2].set_title("FTA Features")
axes[2].set_xlabel("Frequency (Hz)")
axes[2].set_ylabel("FT Imag Amplitude")
axes[2].grid(True)
axes[2].set_xlim(0, 5)

plt.show()

```

```python
def plot_FTA_features(ChToPlot, channel_names,
                      trial_EEG_fft_real_1side,
                      trial_EEG_fft_imaginary_1side,
                      frequency_axis,
                      n_freq_comp,
                      class_ID,
                      data_name,
                    ):

    ch_plot_index = np.where(channel_names==ChToPlot)[0][0]

    FT_real_to_plot = trial_EEG_fft_real_1side[:, ch_plot_index, :]   # (n_trials, n_freq_comp)
    FT_imag_to_plot = trial_EEG_fft_imaginary_1side[:, ch_plot_index, :]

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
    fplot = frequency_axis
    xlim_range = [fplot[0], fplot[n_freq_comp-1]]
    # Auto y-limits
    y_all = FT_real_to_plot.flatten()
    y_lim = np.percentile(np.abs(y_all), 99)
    ylim_range = [-y_lim, y_lim]

    for ax, data_block, ylabel in [(ax1, FT_real_to_plot, "Real"),
                                   (ax2, FT_imag_to_plot, "Imag")]:
        ax.grid(True)
        for class_val, color, label in zip([1,2,3,4], ['r','b','g','m'],
                                           ['cup 1','cup 2','cup 3','cup 4']):
            idx = np.where(class_ID == class_val)[0]

           
            if not np.any(idx):
                continue
            mean_vals = data_block[idx,:].mean(axis=0)
            std_vals  = data_block[idx,:].std(axis=0)
            ax.plot(fplot, mean_vals, color, linewidth=2, label=label)
            ax.plot(fplot, mean_vals + std_vals, f':{color}', linewidth=1)
            ax.plot(fplot, mean_vals - std_vals, f':{color}', linewidth=1)
        ax.set_xlim(xlim_range)
        ax.set_ylim(ylim_range)
        ax.set_ylabel(ylabel)

    ax2.set_xlabel("Frequency (Hz)")
    ax1.set_title(f"{data_name} - {ChToPlot} FTA")
    ax1.legend()
    plt.tight_layout()
    plt.show()


    
```

```python
# Call
plot_FTA_features(
    ChToPlot="C3",
    channel_names=channel_names,
    trial_EEG_fft_real_1side=_real_arr,
    trial_EEG_fft_imaginary_1side=ima_arr,
    frequency_axis=freq_axis,
    n_freq_comp=n_freq_comp,
    class_ID=labels,
    data_name="Reaching_Grasping"
)
```
